```python
# Distribution of content type
content_type_counts = netflix_df['type'].value_counts()

plt.figure(figsize=(8, 5))
content_type_counts.plot(kind='bar', color=['#FF6347', '#4682B4'])
plt.title('Distribution of Content Type')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
# Top 10 countries by content count
top_countries = netflix_df['country'].value_counts().head(10)

plt.figure(figsize=(12, 7))
top_countries.plot(kind='bar', color='#FF6347')
plt.title('Top 10 Countries by Content Count')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```

```python
# Count plot of ratings
plt.figure(figsize=(14, 7))
sns.countplot(x='rating', data=netflix_df, order=netflix_df['rating'].value_counts().index, palette='Set3', hue=netflix_df['rating'], legend=False)
plt.title('Count Plot of Ratings')
plt.xlabel('Count')
plt.ylabel('Rating')

```

```python
# Extracting top 10 directors
top_10_directors = netflix_df['director'].value_counts().head(10).index

# Count plot of top 10 directors
plt.figure(figsize=(14, 7))
sns.countplot(x='director', data=netflix_df[netflix_df['director'].isin(top_10_directors)], palette='Set1', order=top_10_directors, hue='director', legend=False)
plt.title('Count Plot of Top 10 Directors')
plt.xlabel('Director')
plt.ylabel('Count')
plt.show()
```

```python
# Separate movies and TV shows
movies = netflix_df[netflix_df['type'] == 'Movie']
tv_shows = netflix_df[netflix_df['type'] == 'TV Show']
 
# Prepare data for bar chart
movies_duration_per_year = movies.groupby('release_year')['duration'].mean().reset_index()
tv_shows_duration_per_year = tv_shows.groupby('release_year')['duration'].mean().reset_index()

# Bar Chart for Movies and TV Shows Duration
fig_bar = px.bar(
    movies_duration_per_year,
    x='release_year',
    y='duration',
    labels={'duration': 'Average Duration (minutes)', 'release_year': 'Release Year'},
    title='Average Duration of Movies per Year',
    hover_data=['duration']
)
fig_bar.show()
```

```python
# Line Chart for Movies and TV Shows Duration
fig_line = px.line(
    movies_duration_per_year,
    x='release_year',
    y='duration',
    labels={'duration': 'Average Duration (minutes)', 'release_year': 'Release Year'},
    title='Average Duration of Movies per Year'
)
fig_line.show()
```
