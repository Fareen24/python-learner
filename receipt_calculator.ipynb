```python
import datetime

# Store information
store_name = "Chijo Grocery Store"
store_address = "Magadi Rd, Kajiado"
cashier_name = "Ariah Fareen"

# Get the current date and time
now = datetime.datetime.now()
date_time = now.strftime("%d-%m-%Y %H:%M:%S")

# Dictionary to store the products available
fruits_available = {
    1: {"fruit": "Apple", "price": 30},
    2: {"fruit": "Banana", "price": 10},
    3: {"fruit": "Orange", "price": 15},
    4: {"fruit": "Pineapple", "price": 100},
    5: {"fruit": "Mango", "price": 25}
}

# Global variable to store the invoice total
invoice_total = 0

# Function to generate the invoice
def generate_invoice():
    print("Hello our valued customer, welcome!")
    print('Select the fruits you want to buy: ')

    # Display the available fruits and their prices
    for fruit_id, fruit_info in fruits_available.items():
        print(f"{fruit_id} - {fruit_info['fruit']}: ${fruit_info['price']:.2f}")

    cart = {}                                                                    # initializes an empty dictionary to store the customer's selections.
    while True:                                                                  # This starts an infinite loop that will continue until the customer chooses to finish.
        fruit_choices = input("Enter the fruit you want to buy (or 'f' to finish): ")
        if fruit_choices.lower() == 'f':
            break
        if int(fruit_choices) not in fruits_available:
            print("Invalid fruit choice. Please try again.")
            continue

        quantity = int(input(f"How many {fruits_available[int(fruit_choices)]['fruit']}s do you want? "))
        if quantity < 1:
            print("Invalid quantity. Please try again.")
            continue

        cart[int(fruit_choices)] = quantity                                      # adds the selected fruit and quantity to the cart dictionary.

    global invoice_total                                                         # declare the invoice_total variable as global, so it can be accessed and modified within the function.
    invoice_total = 0                                                            # resets the invoice_total to 0 for each new invoice.
    print("\n--- Invoice ---")
    print(f"Store Name: {store_name}")
    print(f"Store Address: {store_address}")
    print(f"Cashier: {cashier_name}")
    print(f"Date and Time: {date_time}")
    print("-------------------")
    '''
    The for loop iterates over the items in the cart dictionary. Then it iterates over the items in the cart dictionary.
    It proceeds to calculate the total cost for the current item.
    Then it adds the item total to the invoice_total prints the invoice total
    '''
    for product_id, quantity in cart.items():
        product = fruits_available[product_id]
        item_total = product['price'] * quantity
        invoice_total += item_total
        print(f"{product['fruit']} x{quantity}: ${item_total:.2f}")

    print(f"\nTotal: ${invoice_total:.2f}")

# Function to process the payment
def process_payment():
    global invoice_total                                                         # declares the invoice_total variable as global, so it can be accessed and modified within the function.
    while True:                                                                  #  starts an infinite loop that will continue until the customer provides a valid payment.
        payment = float(input("\nEnter payment amount: $"))
        if payment >= invoice_total:
            change = payment - invoice_total
            print(f"\nPayment received, Thank you. Change: ${change:.2f}")
            break
        else:
            print("Insufficient payment. Please try again.")

# Run the invoice generator
generate_invoice()
process_payment()









```
